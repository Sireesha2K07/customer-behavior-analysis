```python
import pandas as pd
df = pd.read_csv('/content/ecommerce_customer_data_large.csv')

df.head()
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['Purchase Date']=pd.to_datetime(df['Purchase Date'])
```

```python
df.describe
```

```python
df['Avg Price Per Item']=df['Total Purchase Amount']/df['Quantity']
df['Purchase Month']=df['Purchase Date'].dt.month
```

```python
latest_date=df['Purchase Date'].max()
rfm=df.groupby('Customer ID').agg({
    'Purchase Date':lambda x:(latest_date-x.max()).days,
    'Customer ID': 'count',
    'Total Purchase Amount':'sum'
    })
rfm.columns=['Recency','Frequency','Monetary']
rfm.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
rfm_scaled=scaler.fit_transform(rfm)
```

```python
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=4,random_state=42)
rfm['Cluster']=kmeans.fit_predict(rfm_scaled)
rfm.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(x='Cluster',data=rfm)
plt.title('Customer Segments')
plt.show()
```

```python
monthly_sales=df.groupby('Purchase Month')['Total Purchase Amount'].sum()
monthly_sales.plot(kind='line',marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

```python
sns.countplot(x='Churn',data=df)
plt.title('Churn Distribution')
plt.show()
```

```python
category_sales=df.groupby('Product Category')['Total Purchase Amount'].sum()
category_sales.plot(kind='bar')
plt.title('Sales by Product Category')
plt.ylabel('Sales')
plt.show()
```

```python
rfm.groupby('Cluster').mean()
```

```python
seg_labels={1:'Champions',2:'Loyal',0:'At-Risk',3:'Hibernating'}
rfm['SegmentLabel']=rfm['Cluster'].map(seg_labels)
print(rfm['SegmentLabel'].value_counts())
rfm.head()
```

```python
import matplotlib.pyplot as plt
rfm['SegmentLabel'].value_counts().plot.pie(
    autopct='%1.1f%%',
    figsize=(6,6),
    colors=['gold','steelblue','tomato','grey']
)
plt.title('Customer Segment Distribution')
plt.ylabel('')
plt.axis('equal')
plt.tight_layout()
plt.savefig('seg_pie.png',dpi=150)
plt.show()
```

```python
fig,axes=plt.subplots(1,3,figsize=(15,5))
for ax,col in zip(axes,['Recency','Monetary']):
   rfm.boxplot(column=col,by='SegmentLabel',ax=ax)
   ax.set_title(col)
   ax.set_xlabel('')

plt.suptitle('RFM by Segment')
plt.tight_layout()
plt.savefig('seg_box.png',dpi=150)
plt.show()
```

```python
df['YearMonth']=df['Purchase Date'].dt.to_period('M')
df['TotalSpend']=df['Total Purchase Amount']
```

```python
monthly=df.groupby('YearMonth')['TotalSpend'].sum()
monthly.plot(figsize=(12,4),marker='o',color='steelblue')
plt.title('Monthly Revenue Trend')
plt.ylabel('Total Revenue')
plt.tight_layout()
plt.savefig('monthly_trend.png',dpi=150)
plt.show()
```

```python
cat_rev=df.groupby('Product Category')['TotalSpend'].sum().sort_values()
cat_rev.plot.barh(figsize=(8,4),color='coral')
plt.title('Revenue by Product Category')
plt.tight_layout()
plt.savefig('cat_rev.png',dpi=150)
plt.show()
```

```python
import seaborn as sns
threshold=rfm['Recency'].quantile(0.75)
rfm['ChurnRisk']=rfm['Recency']>threshold
plt.figure(figsize=(8,4))
sns.histplot(rfm['Recency'],bins=30,kde=True,color='tomato')
plt.axvline(threshold,color='black',linestyle='--',label=f'Churn Risk Threshold ({int(threshold)}days)')
plt.title('Recency Distribution - Churn Risk Distribution')
plt.legend()
plt.tight_layout()
plt.savefig('churn_dist.png',dpi=150)
plt.show()
print(f"At-risk customers:{rfm['ChurnRisk'].sum()}")
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
rfm_scaled=scaler.fit_transform(rfm[['Recency','Frequency','Monetary']])

pca=PCA(n_components=2)
components=pca.fit_transform(rfm_scaled)
rfm['PC1']=components[:,0]
rfm['PC2']=components[:,1]
plt.figure(figsize=(8,6))
colors_map={'Champions':'gold','Loyal':'steelblue','At-Risk':'tomato','Hibernating':'grey'}
for seg,grp in rfm.groupby('SegmentLabel'):
  plt.scatter(grp['PC1'],grp['PC2'],label=seg,alpha=0.6,color=colors_map[seg])
plt.legend()
plt.title('Customer Segments - PCA View')
plt.tight_layout()
plt.savefig('pca_clusters.png',dpi=150)
plt.show()
```

```python
from reportlab.platypus import (SimpleDocTemplate, Paragraph, Spacer,
                                 Image, Table, TableStyle, PageBreak)
from reportlab.lib.styles import getSampleStyleSheet
from reportlab.lib.pagesizes import letter
from reportlab.lib import colors

doc = SimpleDocTemplate("Customer_Behavior_Report.pdf", pagesize=letter,
                        rightMargin=50, leftMargin=50,
                        topMargin=50, bottomMargin=50)
styles = getSampleStyleSheet()
story = []

story.append(Spacer(1, 60))
story.append(Paragraph("Customer Behavior Analysis", styles['Title']))
story.append(Paragraph("Alfido Tech — Strategic Insights Report", styles['Heading2']))
story.append(Spacer(1, 10))
story.append(Paragraph("Prepared using E-Commerce Customer Dataset", styles['Normal']))
story.append(Spacer(1,15))
story.append(Paragraph("1. Executive Summary", styles['Heading1']))
story.append(Paragraph(
    "This report analyzes customer transaction data to identify behavioral segments, "
    "purchase patterns, and churn risks for Alfido Tech. Using RFM analysis combined "
    "with K-Means clustering, four distinct customer segments were identified: "
    "Champions, Loyal, At-Risk, and Hibernating customers.",
    styles['Normal']))
story.append(Spacer(1, 15))

story.append(Paragraph("2. Customer Segment Profiles", styles['Heading1']))
story.append(Spacer(1, 8))

seg_data = rfm.groupby('SegmentLabel').agg(
    Count=('Recency','count'),
    Avg_Recency=('Recency','mean'),
    Avg_Frequency=('Frequency','mean'),
    Avg_Monetary=('Monetary','mean')
).round(1).reset_index()

table_data = [['Segment','Count','Avg Recency(days)','Avg Frequency','Avg Spend($)']]
for _, row in seg_data.iterrows():
    table_data.append([
        row['SegmentLabel'], str(row['Count']),
        str(row['Avg_Recency']), str(row['Avg_Frequency']),
        str(row['Avg_Monetary'])
    ])

t = Table(table_data, colWidths=[100,60,110,100,100])
t.setStyle(TableStyle([
    ('BACKGROUND',    (0,0), (-1,0),  colors.HexColor('#2E4057')),
    ('TEXTCOLOR',     (0,0), (-1,0),  colors.white),
    ('FONTNAME',      (0,0), (-1,0),  'Helvetica-Bold'),
    ('ALIGN',         (0,0), (-1,-1), 'CENTER'),
    ('GRID',          (0,0), (-1,-1), 0.5, colors.grey),
    ('ROWBACKGROUNDS',(0,1), (-1,-1), [colors.whitesmoke, colors.white]),
]))
story.append(t)

descriptions = {
    'Champions':   'Bought recently, buy often, spend the most. Reward and retain.',
    'Loyal':       'Buy regularly with good spend. Strong upsell targets.',
    'At-Risk':     'Were active but not purchasing recently. Need win-back campaigns.',
    'Hibernating': 'Last purchase very long ago. Low frequency and spend.'
}
for seg, desc in descriptions.items():
    story.append(Paragraph(f"<b>{seg}:</b> {desc}", styles['Normal']))
    story.append(Spacer(1, 5))

story.append(PageBreak())

story.append(Paragraph("3. Visualizations", styles['Heading1']))
charts = [
    ('seg_pie.png',       'Figure 1: Customer Segment Distribution'),
    ('rfm_box.png',       'Figure 2: RFM Distributions by Segment'),
    ('monthly_trend.png', 'Figure 3: Monthly Revenue Trend'),
    ('cat_rev.png',       'Figure 4: Revenue by Product Category'),
    ('pca_clusters.png',  'Figure 5: PCA Cluster Visualization'),
    ('churn_dist.png',    'Figure 6: Churn Risk — Recency Distribution'),
]
for img_file, caption in charts:
    story.append(Paragraph(caption, styles['Heading3']))
    story.append(Image(img_file, width=440, height=200))
    story.append(Spacer(1, 15))

story.append(PageBreak())

story.append(Paragraph("4. Churn Risk Summary", styles['Heading1']))
churn_count = int(rfm['ChurnRisk'].sum())
total = len(rfm)
pct = round(churn_count / total * 100, 1)
story.append(Paragraph(
    f"Out of {total} customers, {churn_count} ({pct}%) are at high churn risk "
    f"(recency above {int(rfm['Recency'].quantile(0.75))} days). "
    "Immediate re-engagement is recommended.",
    styles['Normal']))
story.append(Paragraph("5. Actionable Recommendations", styles['Heading1']))
story.append(Spacer(1, 10))
recs = [
    ("1. Reward Champions with VIP Programs",
     "Offer exclusive early access, loyalty points, and personalized discounts to top customers."),
    ("2. Win-Back At-Risk Customers",
     "Send automated emails with time-limited 10-15% discounts to customers inactive 60+ days."),
    ("3. Upsell to Loyal Customers",
     "Cross-sell premium bundles in their favourite categories to increase average order value."),
    ("4. Optimize Top Revenue Categories",
     "Increase marketing for Electronics and Clothing. Run promotions on underperforming categories."),
    ("5. Implement Real-Time Churn Alerts",
     "Build a live dashboard flagging customers exceeding the 75th percentile recency threshold."),
]
for title, body in recs:
    story.append(Paragraph(title, styles['Heading3']))
    story.append(Paragraph(body, styles['Normal']))
    story.append(Spacer(1, 12))

doc.build(story)
print("PDF generated successfully!")
```
